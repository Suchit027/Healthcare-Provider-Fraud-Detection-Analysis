```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
# train data
train = pd.read_csv("./archive/Train-1542865627584.csv")
train_inpatient = pd.read_csv("./archive/Train_Inpatientdata-1542865627584.csv")
train_outpatient = pd.read_csv("./archive/Train_Outpatientdata-1542865627584.csv")
train_beneficiary = pd.read_csv("./archive/Train_Beneficiarydata-1542865627584.csv")

# test data
test = pd.read_csv("./archive/Test-1542969243754.csv")
test_inpatient = pd.read_csv("./archive/Test_Inpatientdata-1542969243754.csv")
test_outpatient = pd.read_csv("./archive/Test_Outpatientdata-1542969243754.csv")
test_beneficiary = pd.read_csv("./archive/Test_Beneficiarydata-1542969243754.csv")
```

```python
# Convert 'DOB' and 'DOD' to datetime format
train_beneficiary['DOB'] = pd.to_datetime(train_beneficiary['DOB'], format='%Y-%m-%d')
train_beneficiary['DOD'] = pd.to_datetime(train_beneficiary['DOD'], format='%Y-%m-%d')
test_beneficiary['DOB'] = pd.to_datetime(test_beneficiary['DOB'], format='%Y-%m-%d')
test_beneficiary['DOD'] = pd.to_datetime(test_beneficiary['DOD'], format='%Y-%m-%d')

# Calculate age in years and round to the nearest integer
train_beneficiary['Age'] = round((train_beneficiary['DOD'] - train_beneficiary['DOB']).dt.days / 365)
test_beneficiary['Age'] = round((test_beneficiary['DOD'] - test_beneficiary['DOB']).dt.days / 365)
```

```python
train_beneficiary['DOD'].max()
```

```python
# The final DOD date is '2009-12-01' . We can use that to calculate the ages of those with no DOD

# Convert 'DOB' to datetime format
train_beneficiary['DOB'] = pd.to_datetime(train_beneficiary['DOB'], format='%Y-%m-%d')
test_beneficiary['DOB'] = pd.to_datetime(test_beneficiary['DOB'], format='%Y-%m-%d')

# Calculate age for beneficiaries with missing 'Age' values, assuming 2009-12-01 as the reference date
train_beneficiary['Age'] = train_beneficiary['Age'].fillna(
    round((pd.to_datetime('2009-12-01') - train_beneficiary['DOB']).dt.days / 365)
)
test_beneficiary['Age'] = test_beneficiary['Age'].fillna(
    round((pd.to_datetime('2009-12-01') - test_beneficiary['DOB']).dt.days / 365)
)
```

```python
# Create a new 'AliveorDead' column and flag deceased beneficiaries
train_beneficiary['AliveorDead'] = train_beneficiary['DOD'].notna().astype(int)
test_beneficiary['AliveorDead'] = test_beneficiary['DOD'].notna().astype(int)
```

```python
# Calculate number of days patient was admitted from AdmissionDt and DischargeDt

# Convert 'AdmissionDt' and 'DischargeDt' to datetime format
train_inpatient['AdmissionDt'] = pd.to_datetime(train_inpatient['AdmissionDt'], format='%Y-%m-%d')
train_inpatient['DischargeDt'] = pd.to_datetime(train_inpatient['DischargeDt'], format='%Y-%m-%d')
test_inpatient['AdmissionDt'] = pd.to_datetime(test_inpatient['AdmissionDt'], format='%Y-%m-%d')
test_inpatient['DischargeDt'] = pd.to_datetime(test_inpatient['DischargeDt'], format='%Y-%m-%d')

# Calculate the number of days of admission, adding 1 to include both the admission and discharge days
train_inpatient['NumberofDaysAdmitted'] = (train_inpatient['DischargeDt'] - train_inpatient['AdmissionDt']).dt.days + 1
test_inpatient['NumberofDaysAdmitted'] = (test_inpatient['DischargeDt'] - test_inpatient['AdmissionDt']).dt.days + 1
```

```python
# Convert 'ClaimStartDt' and 'ClaimEndDt' to datetime format
train_inpatient['ClaimEndDt'] = pd.to_datetime(train_inpatient['ClaimEndDt'], format='%Y-%m-%d')
train_inpatient['ClaimStartDt'] = pd.to_datetime(train_inpatient['ClaimStartDt'], format='%Y-%m-%d')

test_inpatient['ClaimEndDt'] = pd.to_datetime(test_inpatient['ClaimEndDt'], format='%Y-%m-%d')
test_inpatient['ClaimStartDt'] = pd.to_datetime(test_inpatient['ClaimStartDt'], format='%Y-%m-%d')

# Duration of Claim
train_inpatient['DurationofClaim'] = (train_inpatient['ClaimEndDt'] - train_inpatient['ClaimStartDt']).dt.days
test_inpatient['DurationofClaim'] = (test_inpatient['ClaimEndDt'] - test_inpatient['ClaimStartDt']).dt.days
```

```python
# Add column admitted

train_inpatient['Admitted'] =1
test_inpatient['Admitted'] =1
```

```python
# Add column admitted

train_outpatient['Admitted'] = 0
test_outpatient['Admitted'] = 0
```

```python
# Convert 'ClaimStartDt' and 'ClaimEndDt' to datetime format
train_outpatient['ClaimEndDt'] = pd.to_datetime(train_outpatient['ClaimEndDt'], format='%Y-%m-%d')
train_outpatient['ClaimStartDt'] = pd.to_datetime(train_outpatient['ClaimStartDt'], format='%Y-%m-%d')

test_outpatient['ClaimEndDt'] = pd.to_datetime(test_outpatient['ClaimEndDt'], format='%Y-%m-%d')
test_outpatient['ClaimStartDt'] = pd.to_datetime(test_outpatient['ClaimStartDt'], format='%Y-%m-%d')

# Duration of Claim
train_outpatient['DurationofClaim'] = (train_outpatient['ClaimEndDt'] - train_outpatient['ClaimStartDt']).dt.days
test_outpatient['DurationofClaim'] = (test_outpatient['ClaimEndDt'] - test_outpatient['ClaimStartDt']).dt.days
```

```python
#Identify common columns between Outpatient and inpatient data

common_cols = list(set(train_inpatient.columns).intersection(set(train_outpatient.columns)))
print(common_cols)
```

```python
# Merge the DataFrames on the common columns, using an outer join

train_all_patient = pd.merge(train_outpatient, train_inpatient, on=common_cols, how='outer')
test_all_patient = pd.merge(test_outpatient, test_inpatient, on=common_cols, how='outer')

print(train_all_patient.shape)
print(test_all_patient.shape)
```

```python
df_train = train_all_patient.merge(train_beneficiary, on='BeneID', how='inner')

df_test = test_all_patient.merge(test_beneficiary, on='BeneID', how='inner')
```

```python
# shape of dataset
print('Training data shape: ', df_train.shape)
print('Test data shape: ', df_test.shape)
```

```python
# Lets merge patient data with fradulent providers details data with "Provider" as joining key for inner join

df_train1 = pd.merge(train, df_train,on='Provider')

df_test1 = pd.merge(test, df_test,on='Provider')
```

```python
# this value indicates whether the beneficiary has renal disease
df_train1['RenalDiseaseIndicator'] = df_train1['RenalDiseaseIndicator'].replace('Y','1')
df_train1['RenalDiseaseIndicator'] = df_train1['RenalDiseaseIndicator'].astype(int)
    
    
df_test1['RenalDiseaseIndicator'] = df_test1['RenalDiseaseIndicator'].replace('Y','1')
df_test1['RenalDiseaseIndicator'] = df_test1['RenalDiseaseIndicator'].astype(int)


# Drop DOD and DOB

df_train1 = df_train1.drop(columns=['DOB', 'DOD'], axis=1)
df_test1 = df_test1.drop(columns=['DOB', 'DOD'], axis=1)
```

```python
df_train1.head()
```

```python
# ClaimDiagnoseIndex, ClmProcedureIndex, ChronicCondIndex

df_train1['ClmDiagnosisCodeIndex'] = df_train1.filter(regex='ClmDiagnosisCode_').notnull().sum(axis=1)
df_test1['ClmDiagnosisCodeIndex'] = df_test1.filter(regex='ClmDiagnosisCode_').notnull().sum(axis=1)

df_train1['ClmProcedureCodeIndex'] = df_train1.filter(regex='ClmProcedureCode_').notnull().sum(axis=1)
df_test1['ClmProcedureCodeIndex'] = df_test1.filter(regex='ClmProcedureCode_').notnull().sum(axis=1)
```

```python
# drop all features starting 

columns_to_drop = df_train1.filter(regex='ClmProcedureCode_|ClmDiagnosisCode_').columns
df_train1 = df_train1.drop(columns_to_drop, axis=1)
df_test1 = df_test1.drop(columns_to_drop, axis=1)
```

```python
# fill nan for NumberofDaysAdmitted with 0

df_train1['NumberofDaysAdmitted'] = df_train1['NumberofDaysAdmitted'].fillna(0)
df_test1['NumberofDaysAdmitted'] = df_test1['NumberofDaysAdmitted'].fillna(0)
```

```python
# df_train1 = df_train1.dropna(subset=['AttendingPhysician'])
# df_test1 = df_test1.dropna(subset=['AttendingPhysician'])

# Fill missing values with mean

df_train1['DeductibleAmtPaid'] = df_train1['DeductibleAmtPaid'].fillna(df_train1['DeductibleAmtPaid'].mean())
df_test1['DeductibleAmtPaid'] = df_test1['DeductibleAmtPaid'].fillna(df_test1['DeductibleAmtPaid'].mean())
```

```python
df_train1.isna().sum()
```

```python
to_drop = ['AttendingPhysician', 'Provider', 'BeneID' ,'ClaimID', 'ClaimStartDt', 'ClaimEndDt', 'AdmissionDt', 'DischargeDt', 'OperatingPhysician', 'OtherPhysician', 'ClmAdmitDiagnosisCode', 'DiagnosisGroupCode']
df_train1 = df_train1.drop(to_drop, axis= 1)
df_test1 = df_test1.drop(to_drop, axis= 1)
```

```python
# convert target to numerical values (0, 1)

df_train1['PotentialFraud'] = df_train1['PotentialFraud'].replace({'No':0, 'Yes': 1})
```

```python
df_train1.head()
```

```python
for x in df_train1.columns:
    df_train1[x] = df_train1[x].astype(float)
```

```python
# Split data into train and validation set

df_train2 , df_val = train_test_split(df_train1, test_size=0.10)

y_train = df_train2.pop('PotentialFraud')
X_train = df_train2

y_val = df_val.pop('PotentialFraud')
X_val = df_val

X_test = df_test1


print(X_train.shape, X_val.shape, X_test.shape)
```

```python
X_train.to_csv('X_train.csv', index= False, header= False)
y_train.to_csv('y_train.csv', index= False, header= False)
X_val.to_csv('X_test.csv', index= False, header= False)
y_val.to_csv('y_test.csv', index= False, header= False)
```
